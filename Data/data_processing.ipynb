```python
from datasets import Dataset, DatasetDict
import json, os

def load_split(folder, difficulty):
    examples = []

    txt_files = sorted([f for f in os.listdir(folder) if f.startswith("problem-") and f.endswith(".txt")])

    for txt_file in txt_files:
        problem_id = txt_file.replace(".txt", "")  

        # match the json filename
        json_file = f"truth-{problem_id}.json"

        txt_path = os.path.join(folder, txt_file)
        json_path = os.path.join(folder, json_file)

        # to check for mismatches
        if not os.path.exists(json_path):
            print(f"Warning: JSON missing for {txt_file}")
            continue

        # read sentences per line
        with open(txt_path, encoding="utf-8") as f:
            sentences = [line.strip() for line in f.readlines() if line.strip()]

        # read truth file
        with open(json_path, encoding="utf-8") as f:
            truth = json.load(f)

        examples.append({
            "id": problem_id,                       # "problem-x"
            "difficulty": difficulty,
            "text": sentences,                      # list of sentences
            "authors": truth["authors"],            # integer of num_authors
            "changes": truth["changes"],            # binary list of author changes
        })

    return Dataset.from_list(examples)
```

```python
dataset = DatasetDict({
    "train_easy": load_split("easy/train", "easy"),
    "valid_easy": load_split("easy/validation", "easy"),
    "train_medium": load_split("medium/train", "medium"),
    "valid_medium": load_split("medium/validation", "medium"),
    "train_hard": load_split("hard/train", "hard"),
    "valid_hard": load_split("hard/validation", "hard"),
})
```

```python
dataset.save_to_disk("multiauthor_dataset/")
```
