### Split dataset to bypass repo upload limit

```python
from datasets import load_from_disk

# Load the dataset
dataset = load_from_disk("tokenized_train_dataset")

# Compute split index
mid = len(dataset) // 2

# Split into two
shard1 = dataset.select(range(0, mid))
shard2 = dataset.select(range(mid, len(dataset)))

# Save shards
shard1.save_to_disk("tokenized_train_dataset_shard_0")
shard2.save_to_disk("tokenized_train_dataset_shard_1")
```

### Merge them again (run this if loaded from git)

```python
from datasets import load_from_disk, concatenate_datasets

# Load the two shards
shard1 = load_from_disk("Data/tokenized_train_dataset_shard_0")
shard2 = load_from_disk("Data/tokenized_train_dataset_shard_1")

# Concatenate
full_dataset = concatenate_datasets([shard1, shard2])

# Use it like normal
print(full_dataset)

full_dataset.save_to_disk("tokenized_train_dataset")
```
